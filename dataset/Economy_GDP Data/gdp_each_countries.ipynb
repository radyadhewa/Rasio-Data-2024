```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("27e2a688-13f8-4852-b6c0-0caaed9bbbe9_Series - Metadata.csv")
df.head()
```

```python
filtered_df = df[df['Series Name']=='GDP (constant 2015 US$)']
filtered_df
```

```python
filtered_df.drop(filtered_df.index[-1], inplace=True)
```

```python
filtered_df.info()
```

```python
# Step 1: Filter for 'GDP (constant 2015 US$)' rows
gdp_df = filtered_df[filtered_df['Series Name'] == 'GDP (constant 2015 US$)']

# Step 2: Melt the DataFrame to make 'Year' a single column
melted_df = gdp_df.melt(id_vars=['Country Name'], 
                        value_vars=['2012 [YR2012]', '2013 [YR2013]', '2014 [YR2014]', '2015 [YR2015]', 
                                    '2016 [YR2016]', '2017 [YR2017]', '2018 [YR2018]', '2019 [YR2019]', 
                                    '2020 [YR2020]', '2021 [YR2021]', '2022 [YR2022]', '2023 [YR2023]'], 
                        var_name='Year', 
                        value_name='GDP')

# Step 3: Clean up the 'Year' column to extract just the year number
melted_df['Year'] = melted_df['Year'].str.extract(r'(\d{4})')

# Step 4: Pivot the DataFrame to make 'Country Name' as columns
final_df = melted_df.pivot(index='Year', columns='Country Name', values='GDP')

# Step 5: Rename the columns to include 'GDP' before each country name
final_df.columns = ['GDP ' + col for col in final_df.columns]
```

```python
final_df
```

```python
final_df.to_csv("all_gdp.csv")
```
